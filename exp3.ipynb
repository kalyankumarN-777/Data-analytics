```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

df = pd.read_csv("F:\\Codding files\\.vscode\\python\\DA lab\\car_price_dataset.csv")
print("Original DataFrame:\n", df)

dataf = df.describe()
print(dataf)

plt.boxplot(df['Price'])  
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Price Distribution for Civic')
plt.legend() 
plt.show()
```

```python
sns.violinplot(y=df['Price']) 
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Price Distribution for Civic')
plt.show()
```

```python
plt.scatter(df['Price'], df['Mileage']) 
plt.xlabel('Price')
plt.ylabel('Mileage')
plt.title('Price vs Mileage')
plt.show()
```

```python
plt.stackplot(df.index, df['Price'], df['Mileage'], labels=['Price', 'Mileage'])  
plt.xlabel('Index') 
plt.ylabel('Values')
plt.title('Stack Plot of Price and Mileage')
plt.legend(loc='upper left')
plt.show()
```

```python
plt.quiver(df['Price'], df['Year'], df['Mileage'], df['Engine_Size'], scale=1) 
plt.xlabel('Price')
plt.ylabel('Mileage')
plt.title('Quiver Plot')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from scipy import stats as st
df = pd.read_csv("F:\\Codding files\\.vscode\\python\\DA lab\\car_price_dataset.csv")
a=df.describe()
print(a)
```

```python
numeric_df = df.select_dtypes(include=np.number)
gmean = st.gmean(numeric_df)
print("Geometric mean of each numeric column:", gmean)
```

```python
numeric_df = df.select_dtypes(include=np.number)
hmean = st.hmean(numeric_df)
print("harmonic mean of each numeric column:", hmean)
```

```python
numeric_df = df.select_dtypes(include=np.number)
Z = st.zscore(numeric_df)
print("Z score of each numeric column:", Z)
```

```python
df['Cumulative Frequency'] = df['Year'].rank(method='max').astype(int)
df_sorted = df.sort_values(by='Year')
plt.figure(figsize=(10, 6))
sns.lineplot(x=df_sorted['Year'], y=df_sorted['Cumulative Frequency'], marker='o')
plt.title('Cumulative Frequency of Car Prices')
plt.xlabel('Year')
plt.ylabel('Cumulative Frequency')
plt.grid(True)
plt.show()
```
