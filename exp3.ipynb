```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
df = pd.read_csv("F:\\Codding files\\.vscode\\python\\DA lab\\car_price_dataset.csv")
print("Original DataFrame:\n",df)
dataf=df.describe()
print(dataf)
```

```python
plt.figure(figsize=(10, 5))
plt.bar(df['Model'], df['Price'])
plt.xlabel('Car Model')
plt.ylabel('Price')
plt.title('Car Prices by Model')
plt.xticks(rotation=90) 
plt.show()
```

```python
civic_prices = df[df['Model'] == 'Civic']['Price']
plt.figure(figsize=(10, 5))
plt.hist(civic_prices, bins=10, alpha=0.7, label='Civic')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Price Distribution for Civic')
plt.legend()
plt.show()
```
